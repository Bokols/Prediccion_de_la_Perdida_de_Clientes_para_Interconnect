#### 3.2.5 ¬øC√≥mo es la evoluci√≥n de los contratos a lo largo del tiempo?

- Dividir los datos en contratos activos y contratos inactivos (o finalizados) y contar cada a√±o nos permite entender c√≥mo var√≠a el n√∫mero de contratos a lo largo del tiempo.

```python
# Contratos activos
active_contracts = df_merged[df_merged['end_date'].isnull()]

# Contratos inactivos
finished_contracts = df_merged[df_merged['end_date'].notna()]

# A√±os ordenados
years = sorted(df_merged['begin_date'].dt.year.unique())

# Inicializaci√≥n de contadores
active_counts = []
finished_counts = []

# Bucle
for year in years:
    # Nuevos contratos activos
    active_year = active_contracts[active_contracts['begin_date'].dt.year == year].shape[0]

    # Contratos finalizados
    finished_year = finished_contracts[finished_contracts['end_date'].dt.year == year].shape[0]

    # A√±adir
    active_counts.append(active_year)
    finished_counts.append(finished_year)

# Imprimir
print(active_counts)
print(finished_counts)
```

```python
# Tama√±o y configuraci√≥n del gr√°fico
bar_width = 0.35
index = np.arange(len(years))
plt.figure(figsize=(10, 4))

# Barras de contratos activos y finalizados
bars1 = plt.bar(index - bar_width/2, active_counts, bar_width,
                label='Active contracts', color='green', alpha=0.7)
bars2 = plt.bar(index + bar_width/2, finished_counts, bar_width,
                label='Inactive contracts', color='red', alpha=0.7)

# T√≠tulo
plt.title('Recuento de contratos activos e inactivos por a√±o')
plt.xlabel('A√±o')
plt.ylabel('N√∫mero de contratos')
plt.xticks(index, years)  # Etiquetas del eje X
plt.legend()

# Contar
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        if height != 0: #Cuente solo si el valor es diferente de 0
            plt.gca().annotate(format(height, '.0f'),
                               (bar.get_x() + bar.get_width() / 2., height),
                               ha='center', va='center', xytext=(0, -20),
                               textcoords='offset points', fontsize=10)

# Imprimir
plt.tight_layout()
plt.show()
```

- Los Contratos Nuevos Activos comienzan en 0 en el primer a√±o, aumentan significativamente en el segundo a√±o (1246) y luego disminuyen gradualmente en los a√±os siguientes. Hay un ligero aumento del quinto a√±o (646) al s√©ptimo a√±o (976), pero cae dr√°sticamente en el octavo y √∫ltimo a√±o de los datos (233).
- Por otro lado, los Contratos Inactivos/Cancelados no tienen registros durante los primeros seis a√±os, lo que indica una retenci√≥n completa durante ese per√≠odo. Sin embargo, a partir del s√©ptimo a√±o, hay un aumento significativo en los contratos inactivos o cancelaciones, con 1409 registros en el s√©ptimo a√±o y 460 en el octavo a√±o.
- Comprender estas fluctuaciones ayuda a identificar posibles cambios en la adquisici√≥n de nuevos clientes o la retenci√≥n de los existentes a lo largo de los a√±os. El aumento repentino de las cancelaciones en el s√©ptimo a√±o puede deberse a varios factores como la insatisfacci√≥n del cliente, los cambios en el mercado o la competencia.

#### 4.4.1 Sobremuestreo
- Implica aumentar la cantidad de ejemplos en la clase minoritaria para que coincida con la cantidad de ejemplos en la clase mayoritaria.
- Este m√©todo no produce p√©rdida de datos; puede ser √∫til cuando hay suficiente tiempo de procesamiento y cuando la clase minoritaria contiene informaci√≥n valiosa que podr√≠a perderse con un submuestreo.
- Por otro lado, puede generar tiempos de entrenamiento m√°s prolongados y aumentar el riesgo de sobreajuste.

```python
# Funci√≥n
def upsample(features, target):
    features_zeros = features[target == False]
    features_ones = features[target == True]
    target_zeros = target[target == False]
    target_ones = target[target == True]

    # N√∫mero de repeticiones
    repeat_times = len(target_zeros) // len(target_ones)

    # Muestras adicionales
    additional_samples = len(target_zeros) % len(target_ones)

    # Aumentar la muestra de la clase minoritaria
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat_times + [
                                   features_ones.sample(additional_samples, replace=True)])
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat_times + [
                                 target_ones.sample(additional_samples, replace=True)])

    # Mezclar datos
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)

    # Devolver
    return features_upsampled, target_upsampled

# Aplicar sobremuestreo
features_upsampled, target_upsampled = upsample(features_train, target_train)
```

```python
# Calcular distribuci√≥n
class_frequency_upsampled = target_upsampled.value_counts(normalize=True)

# Imprimir
print("Class distribution after oversampling:")
print(class_frequency_upsampled)

# Tama√±o del gr√°fico
plt.figure(figsize=(4, 3))

# Gr√°fico
class_frequency_upsampled.plot(kind='bar', color=['gray', 'lightgreen'])

# T√≠tulo
plt.title('Class Distribution After Oversampling')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['False', 'True'])

# Imprimir
plt.show()
```

#### 4.4.2 Submuestreo
- Implica reducir la cantidad de ejemplos en la clase mayoritaria para que coincida con la cantidad de ejemplos en la clase minoritaria.
- Al reducir el tama√±o del conjunto de datos, se puede acelerar el entrenamiento del modelo y mejorar el rendimiento al evitar el sobreajuste a la clase mayoritaria.
- Por otro lado, se puede perder informaci√≥n valiosa al eliminar ejemplos de la clase mayoritaria, lo que potencialmente da como resultado un modelo menos preciso.

```python
# Funci√≥n
def downsample(features, target):

    # Separar caracter√≠sticas
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

   
    # Submuestrear la clase mayoritaria (1)
    features_majority_downsampled = features_ones.sample(
        len(target_zeros), random_state=12345, replace=False)
    target_majority_downsampled = target_ones.sample(
        len(target_zeros), random_state=12345, replace=False)
    features_minority = features_zeros
    target_minority = target_zeros

    # Submuestrear la clase mayoritaria
    features_downsampled = pd.concat(
        [features_majority_downsampled, features_minority])
    target_downsampled = pd.concat(
        [target_majority_downsampled, target_minority])

    # Mezclar datos
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)

    # Devolver
    return features_downsampled, target_downsampled


# Aplicar reducci√≥n de muestreo
features_downsampled, target_downsampled = downsample(
    features_train, target_train)
```

```python
# Calcular distribuci√≥n
class_frequency_downsampled = target_downsampled.value_counts(normalize=True)

# Imprimir
print("Class distribution after undersampling:")
print(class_frequency_downsampled)

# Tama√±o del gr√°fico
plt.figure(figsize=(4, 3))

# Gr√°fico
plt.bar(class_frequency_downsampled.index.astype(str),
        class_frequency_downsampled.values, color=['gray', 'lightgreen'])

# T√≠tulo
plt.title('Class Distribution After Undersampling')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.xticks([0, 1], ['False', 'True'])

# Imprimir
plt.tight_layout()
plt.show()
```
